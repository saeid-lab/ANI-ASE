```python
import pandas as pd
```

```python
QM_data = pd.read_csv('IR_Cleaned.txt', sep=' ')
```

```python
QM_data.columns = ['Freq_QM','IR']
```

```python
QM_data
```

```python
QM_data.sort_values(by=['IR'],ascending=False)
```

```python
ML_data = pd.read_csv('IR_ML.txt')
```

```python
result= pd.concat([QM_data,ML_data],axis=1)
```

```python
result
```

```python
result.sort_values(by=['IR'],ascending=False,inplace=True)
```

```python
result
```

```python
def SE(a, b): 
    return (a - b)**2

result['SE'] = result.apply(lambda row: SE(row['Freq_QM'],row['ML']),axis=1)
```

```python
result
```

```python
result.mean(axis=0)
```
